```python
#! /usr/bin/env python
# 03/07/2022 Zheting
from pymatgen.core import Lattice, Structure, Molecule
```

```python
# Read a POSCAR and write to a CIF.
structure = Structure.from_file("/mnt/d/ubuntu/plot/singlelayer/noFe/POSCAR")
# structure.to(filename="CsCl.cif")

# Read an xyz file and write to a Gaussian Input file.
# methane = Molecule.from_file("methane.xyz")
# methane.to(filename="methane.gjf")
print(structure)
```

```python
# Make a supercell
structure.make_supercell([2, 2, 1])
print(structure)
```

```python
structure.append("Fe", [0.00, 0.00, 0.84])
structure.append("Fe", [0.50, 0.50, 0.84])
print(structure)
```

```python
# Get a primitive version of the Structure
structure_2Cu = Structure.get_primitive_structure(structure)
print(structure_2Cu)
```

```python
structure_2Cu.to(filename="testdata/POSCAR_60atom")
```
